```python
## CREATE X and y
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import cross_validate
```

```python
model = Ridge(alpha=100)
```

```python
# SCORING OPTIONS:
# https://scikit-learn.org/stable/modules/model_evaluation.html
scores = cross_validate(model,X_train,y_train,
                         scoring=['neg_mean_absolute_error','neg_mean_squared_error'],cv=10)
```

```python
scores = pd.DataFrame(scores)
```

```python
scores.mean()
```

```python
model = Ridge(alpha=1)
```

```python
# SCORING OPTIONS:
# https://scikit-learn.org/stable/modules/model_evaluation.html
scores = cross_validate(model,X_train,y_train,
                         scoring=['neg_mean_absolute_error','neg_mean_squared_error'],cv=10)
```

```python
scores = pd.DataFrame(scores)
```

```python
scores.mean()
```

```python
model.fit(X_train,y_train)
```

```python
y_final_pred = model.predict(X_test)
```

```python
mean_squared_error(y_test,y_final_test_pred)
```

```python
base_elastic_net_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],'l1_ratio':[.1,.5,.7,.95,.99,1]}
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,
                         param_grid=param_grid,
                         scoring='neg_mean_squared_error',cv=5,verbose=2)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```
