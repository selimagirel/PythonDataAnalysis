```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue_order='level of education')
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
sns.distplot(df,x='salary')
```

```python
df['division'].value_counts()
```

```python
sns.countplot(data=df,x='division')
```

```python
sns.barplot(data=df,x='level of education',y='salary',estimator=np.mean,ci='sd')
```
