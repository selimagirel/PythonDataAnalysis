```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

lines = ax.plot(x,x+1,color='#14fa2f',lw=6,marker='o')
lines[0].set_dashes([5,2,10,2])

```
